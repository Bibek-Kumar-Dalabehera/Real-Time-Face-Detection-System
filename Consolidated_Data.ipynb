```python
import cv2
import os
import pickle
import numpy as np

data_dir = os.path.join(os.getcwd(),'clean data')# connnect to clean data directory
img_dir = os.path.join(os.getcwd(),'images')# connnect to images directory


image_data = []
labels = []

for i in os.listdir(img_dir): # it take each data of images directory
    image = cv2.imread(os.path.join(img_dir,i))#Read each images 
    image = cv2.resize(image,(100,100))# Resize the image
    image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)# we convert to gray color
    image_data.append(image)#append to image list
    labels.append(str(i).split("_")[0])# it split name and no.jpg and add the name to labels list

# convert to eahh image to matrix 
image_data = np.array(image_data)
#add all image name to array
labels = np.array(labels) 


import matplotlib.pyplot as plt
plt.imshow(image_data[56],cmap="gray")
plt.show()

#adding the file and name to coresponding file 
with open(os.path.join(data_dir,"images.p"),'wb') as f:
    pickle.dump(image_data,f)
    
with open(os.path.join(data_dir,"labels.p"),'wb') as f:
    pickle.dump(labels,f)
    
```
